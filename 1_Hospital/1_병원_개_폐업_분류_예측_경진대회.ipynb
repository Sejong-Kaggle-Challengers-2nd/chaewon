#Data Import

```
import pandas as pd
#inst_id를 index열로 넣기
train=pd.read_csv("train.csv",index_col=0)
test=pd.read_csv("test.csv",index_col=0)
submit=pd.read_csv("submission_sample.csv")
```

```
train.head()
```

```
#NaN 을 포함한 행 삭제
train=train.dropna(axis=0)

#0만 있는 행 삭제
train=train.drop(columns=['receivableL1','receivableL2'])
```

```
from sklearn.preprocessing import LabelEncoder
classle=LabelEncoder()
train['OC']=classle.fit_transform(train['OC'].values)
train['sido']=classle.fit_transform(train['sido'].values)
train['instkind']=classle.fit_transform(train['instkind'].values)
train['ownerChange']=classle.fit_transform(train['ownerChange'].values)
train
```

```
y_train=train['OC']
x_train=train.iloc[:,1:]
```

```
x_train.head()
```

```
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test=train_test_split(x_train,y_train,test_size=0.1, random_state=1,stratify=y_train)
```

#모델학습

##Logistic Regression

```
from sklearn.linear_model import LogisticRegression
lg=LogisticRegression(random_state=1)
lg.fit(x_train,y_train)
y_train_pred=lg.predict(x_train)
y_test_pred=lg.predict(x_test)
```

```
from sklearn.metrics import accuracy_score
print(accuracy_score(y_train,y_train_pred))
print(accuracy_score(y_test,y_test_pred))
```

##KNN

```
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(x_train,y_train)
y_train_pred=knn.predict(x_train)
y_test_pred=knn.predict(x_test)
```

```
print(accuracy_score(y_train,y_train_pred))
print(accuracy_score(y_test,y_test_pred))
```

##LDA

```
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda=LinearDiscriminantAnalysis(store_covariance=True)
```

```
lda.fit(x_train,y_train)
y_train_pred=lda.predict(x_train)
y_test_pred=lda.predict(x_test)
```

```
print(accuracy_score(y_train,y_train_pred))
print(accuracy_score(y_test,y_test_pred))
```

##QDA

```
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
qda=QuadraticDiscriminantAnalysis()
qda.fit(x_train,y_train)
y_train_pred=qda.predict(x_train)
y_test_pred=qda.predict(x_test)
```

```
print(accuracy_score(y_train,y_train_pred))
print(accuracy_score(y_test,y_test_pred))
```

##Decision Tree

```
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier(max_depth=5)
dtc.fit(x_train,y_train)
y_train_pred=dtc.predict(x_train)
y_test_pred=dtc.predict(x_test)
```

```
print(accuracy_score(y_train,y_train_pred))
print(accuracy_score(y_test,y_test_pred))
```
